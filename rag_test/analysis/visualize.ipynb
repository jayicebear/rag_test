```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

#  
data = {
    'Method': ['BM25', 'BM25 + Reranker', 'Hybrid', 'Hybrid + Reranker'],
    'Recall@1': [0.118, 0.255, 0.245, 0.343],
    'Recall@3': [0.275, 0.382, 0.431, 0.49],
    'Recall@5': [0.343, 0.441, 0.48, 0.559],
    'MRR': [0.217, 0.329, 0.36, 0.432]
}

df = pd.DataFrame(data)

# 2 subplot 
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# === : BM25 vs BM25 + Reranker ===
df_bm25 = df[df['Method'].str.contains('BM25')].copy()
df_bm25_melted = df_bm25.melt(id_vars='Method', var_name='Metric', value_name='Score')

palette_bm25 = {
    'BM25': '#a1a1a1',
    'BM25 + Reranker': '#0004f1'
}

ax1 = axes[0]
sns.barplot(data=df_bm25_melted, x='Metric', y='Score', hue='Method', 
            palette=palette_bm25, ax=ax1)

# BM25    
metrics = ['Recall@1', 'Recall@3', 'Recall@5', 'MRR']
for i, metric in enumerate(metrics):
    bm25_score = df.loc[df['Method'] == 'BM25', metric].values[0]
    bm25_rerank_score = df.loc[df['Method'] == 'BM25 + Reranker', metric].values[0]
    improvement = ((bm25_rerank_score - bm25_score) / bm25_score) * 100
    
    mid_y = (bm25_score + bm25_rerank_score) / 2
    ax1.annotate('', xy=(i, bm25_rerank_score), xytext=(i, bm25_score),
                arrowprops=dict(arrowstyle='->', color='#e74c3c', lw=2.5))
    ax1.text(i + 0.15, mid_y, f'+{improvement:.1f}%', 
            color='#e01e09', fontweight='bold', fontsize=11, 
            va='center', bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.8))

ax1.set_title('BM25 with/without Reranking', fontsize=14, fontweight='bold', pad=15)
ax1.set_ylabel('Score', fontsize=12)
ax1.set_xlabel('Metric', fontsize=12)
ax1.set_ylim(0, 0.6)
ax1.legend(title='Method', fontsize=10, title_fontsize=11, loc='upper left')
ax1.grid(axis='y', alpha=0.3, linestyle='--')

# === : Hybrid vs Hybrid + Reranker ===
df_hybrid = df[df['Method'].str.contains('Hybrid')].copy()
df_hybrid_melted = df_hybrid.melt(id_vars='Method', var_name='Metric', value_name='Score')

palette_hybrid = {
    'Hybrid': "#a1a1a1",
    'Hybrid + Reranker': '#0004f1'
}

ax2 = axes[1]
sns.barplot(data=df_hybrid_melted, x='Metric', y='Score', hue='Method', 
            palette=palette_hybrid, ax=ax2)

# Hybrid    
for i, metric in enumerate(metrics):
    hybrid_score = df.loc[df['Method'] == 'Hybrid', metric].values[0]
    hybrid_rerank_score = df.loc[df['Method'] == 'Hybrid + Reranker', metric].values[0]
    improvement = ((hybrid_rerank_score - hybrid_score) / hybrid_score) * 100
    
    mid_y = (hybrid_score + hybrid_rerank_score) / 2
    ax2.annotate('', xy=(i, hybrid_rerank_score), xytext=(i, hybrid_score),
                arrowprops=dict(arrowstyle='->', color='#e74c3c', lw=2.5))
    ax2.text(i + 0.15, mid_y, f'+{improvement:.1f}%', 
            color='#e01e09', fontweight='bold', fontsize=11, 
            va='center', bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.8))

ax2.set_title('Hybrid with/without Reranking', fontsize=14, fontweight='bold', pad=15)
ax2.set_ylabel('Score', fontsize=12)
ax2.set_xlabel('Metric', fontsize=12)
ax2.set_ylim(0, 0.6)
ax2.legend(title='Method', fontsize=10, title_fontsize=11, loc='upper left')
ax2.grid(axis='y', alpha=0.3, linestyle='--')

#  
fig.suptitle('Impact of Reranking on Retrieval Performance', 
             fontsize=16, fontweight='bold', y=0.98)

plt.tight_layout()
plt.show()
```
